# Credit Card Fraud Detection


# Abstract

On the era of online transaction and e-commerse fraudulent transactions has become a major problem, as all transactions can easily be completed online by only entering your credit card information. Organizations, consumers, banks, and merchants are put at risk when a data breach leads to monetary theft and ultimately the loss of customers’ loyalty along with the company’s reputation. It is important that creditcard companies are able to recognize fraudlent transactions.

Machine learning had played an imperative role in the detection of credit card fraud in online transactions. Credit card fraud detection is challenging , because the profiles of normal and fraudulent behaviours change constantly and credit card fraud data sets are highly skewed.For predicting these transactions banks make use of various machine learning methodologies, past data has been collected and new features are been used for enhancing the predictive power. The performance of fraud detecting in credit card transactions is greatly affected by the sampling approach on data-set, selection of variables and detection techniques used. This paper investigates the performance of logistic regression, decision tree and random forest for credit card fraud detection. 
Comparison are made for different machine learning algorithms such as Logistic Regression, Decision Trees, Random Forest, to determine which algorithm suits best and can be adapted by credit card merchants for identifying fraud transactions.

Algorithm steps:

Step 1: Read the dataset. 
Step 2: Random Sampling is done on the data set to make it balanced.
Step 3: Divide the dataset into Train dataset and Test dataset.
Step 4: Data is scaled.
Step 5: Accuracy and performance metrics has been calculated to know the efficiency for different algorithms. 
Step6: Then retrieve the best algorithm based on efficiency for the given dataset.

# Data Preprocessing

There is a huge differenc in minimum and maximum of Amount. we need to scale the data to adjust the range.

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
amount=df_c['Amount'].values
df_c['Amount']=scale.fit_transform(amount.reshape(-1, 1))
```

# Holdout validation aproch- train and test split

```python
from sklearn.model_selection import train_test_split
x=df_c.drop(['Class'],axis=True).values
y=df_c['Class']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state = 42)
```

### Evaluation model
Model Evaluation is an essential part of the model development process. 
It helps to find the best model that represents our data and how well the selected model will work in the future.

Trained Data and Testing Data is represented in a confusion matrix which portrays: 
 TP: True Positive which denotes the real data where customers are subjected to fraud and are used for training and were 
    accurately predicted.
 TN: True Negative denotes the data which was not predicted and doesn’t match with the data which was subjected to the fraud.
 FP: False Positive is predicted but there is no possibility of the data to be subjected to the fraud.
 FN: False Negative is not predicted but there is an actual possibility of the data who is subjected to fraud

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf,x_train,y_train,x_test,y_test,train=True):
    if train:
        
        pred=clf.predict(x_train)
        cls_report=pd.DataFrame(classification_report(y_train,pred,output_dict=True))
        print("Train Result:\n======================================")
        print(f"accuracy:{accuracy_score(y_train,pred) *100:.2f}%")
        print("------------------------------------------------------")
        print(f"classification report:\n {cls_report}")
        print("-------------------------------------------------------")
        print(f"Confusion matrix:\n{confusion_matrix(y_train,pred)}\n")
        LABELS = ['Normal', 'Fraud']
        conf_matrix = confusion_matrix(y_train,pred)
        plt.figure(figsize =(4, 4))
        sns.heatmap(conf_matrix, xticklabels = LABELS, 
            yticklabels = LABELS, annot = True, fmt ="d");
        plt.title("Confusion matrix")
        plt.ylabel('True class')
        plt.xlabel('Predicted class')
        plt.show()
    elif train==False:
        pred=clf.predict(x_test)
        cls_report=pd.DataFrame(classification_report(y_test,pred,output_dict=True))
        print("Test Result:\n======================================")
        print(f"accuracy:{accuracy_score(y_test,pred) *100:.2f}%")
        print("------------------------------------------------------")
        print(f"classification report:\n {cls_report}")
        print("-------------------------------------------------------")
        print(f"Confusion matrix:\n{confusion_matrix(y_test,pred)}\n")
        LABELS = ['Normal', 'Fraud']
        conf_matrix = confusion_matrix(y_test,pred)
        plt.figure(figsize =(4, 4))
        sns.heatmap(conf_matrix, xticklabels = LABELS, 
            yticklabels = LABELS, annot = True, fmt ="d");
        plt.title("Confusion matrix")
        plt.ylabel('True class')
        plt.xlabel('Predicted class')
        plt.show()
        
                    
```

### LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr_ypred=lr.predict(x_test)
print_score(lr, x_train, y_train, x_test, y_test, train=True)
print_score(lr, x_train, y_train, x_test, y_test, train=False)
```

Evaluate the confusion matrix, Here, the first row represents positive and the second row represents negative. So, we have 169850 as true positive and 85 are false positive. That says, out of 169850+81=169931, we have 169850 that are successfully classified as a normal transaction and 81 were falsely classified as normal — but they were fraudulent and 219 successfully classified as fraudulent. Since it is a fraud detection our aim is to reduce the false negative.

# DECITION TREE

```python
from sklearn.tree import DecisionTreeClassifier
dtr=DecisionTreeClassifier(max_depth=4)
dtr.fit(x_train,y_train)
dtr_ypred=dtr.predict(x_test)
print("DECITION TREE CLASSIFIER")
print_score(dtr, x_train, y_train, x_test, y_test, train=True)
print_score(dtr, x_train, y_train, x_test, y_test, train=False)
```

## DECISION TREE CLASSIFIER WITH ENTROPY

```python
from sklearn.tree import DecisionTreeClassifier
dtr_entr=DecisionTreeClassifier(max_depth=4,criterion='entropy')
dtr_entr.fit(x_train,y_train)
print("DECITION TREE CLASSIFIER WITH ENTROPY")
print("====================================================")
print_score(dtr_entr, x_train, y_train, x_test, y_test, train=True)
print_score(dtr_entr, x_train, y_train, x_test, y_test, train=False)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtr_gini=DecisionTreeClassifier(max_depth=4,criterion='gini')
dtr_gini.fit(x_train,y_train)
print("DECITION TREE CLASSIFIER WITH GINI")
print("====================================================")
print_score(dtr_gini, x_train, y_train, x_test, y_test, train=True)
print_score(dtr_gini, x_train, y_train, x_test, y_test, train=False)
```

```python
#aucroc
from sklearn.metrics import roc_curve, auc
probs_dtre_test = dtr_entr.predict_proba(x_test)[:, 1] 
probs_dtre_train = dtr_entr.predict_proba(x_train)[:, 1] 
train_fpr, train_tpr, tr_thresholds = roc_curve(y_train, probs_dtre_train)
test_fpr, test_tpr, te_thresholds = roc_curve(y_test, probs_dtre_test)

plt.grid()

plt.plot(train_fpr, train_tpr, label=" AUC TRAIN ="+str(auc(train_fpr, train_tpr)))
plt.plot(test_fpr, test_tpr, label=" AUC TEST ="+str(auc(test_fpr, test_tpr)))
plt.plot([0,1],[0,1],'g--')
plt.legend()
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("AUC(ROC curve) FOR DECISION TREE CLASSIFIER")
plt.grid(color='black', linestyle='-', linewidth=0.5)
plt.show()
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth = 4)
rf.fit(x_train, y_train)
rf_ypred = rf.predict(x_test)
rf_ypredtrain=rf.predict(x_train)
print("RANDOM FOREST CLASSIFIER")
print_score(rf, x_train, y_train, x_test, y_test, train=True)
print_score(rf, x_train, y_train, x_test, y_test, train=False)
```

```python
#rocauc
from sklearn.metrics import roc_curve, auc
probs_rf_test = rf.predict_proba(x_test)[:, 1] 
probs_rf_train = rf.predict_proba(x_train)[:, 1] 
train_fpr, train_tpr, tr_thresholds = roc_curve(y_train, probs_rf_train)
test_fpr, test_tpr, te_thresholds = roc_curve(y_test, probs_rf_test)

plt.grid()

plt.plot(train_fpr, train_tpr, label=" AUC TRAIN ="+str(auc(train_fpr, train_tpr)))
plt.plot(test_fpr, test_tpr, label=" AUC TEST ="+str(auc(test_fpr, test_tpr)))
plt.plot([0,1],[0,1],'g--')
plt.legend()
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("AUC(ROC curve) FOR RANDOM FOREST")
plt.grid(color='black', linestyle='-', linewidth=0.5)
plt.show()
```
